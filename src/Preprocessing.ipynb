# Resize Images

```python
import cv2
import os

# Path menuju folder input dan output
input_folder = "../dataset/raw/Induk aksara"
output_folder = "../dataset/wo aug/preprocessed/Induk aksara"

# Loop melalui setiap file dalam folder input
for root, dirs, files in os.walk(input_folder):
    for file in files:

        # Baca gambar menggunakan OpenCV
        img_path = os.path.join(root, file)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)

        # Resize gambar menjadi ukuran 224x224 pixel
        resize_img = cv2.resize(img, (224, 224))

        # Tentukan output path untuk menyimpan gambar yang sudah diresize
        folder_name = os.path.relpath(root, input_folder)
        output_path = os.path.join(output_folder, folder_name)

        # Buat direktori output jika belum ada
        if not os.path.exists(output_path):
            os.makedirs(output_path)
            
        # Simpan gambar yang sudah diresize
        cv2.imwrite(os.path.join(output_path, file), resize_img)
```

# Rename Folders and Files

```python
import os
import shutil

# Path to your directory
mypath = "../dataset/preprocessed/Induk aksara"

# Fungsi untuk mengubah nama file atau folder
def rename_files_folders(path):
    for root, dirs, files in os.walk(path):
        for name in dirs:
            if "E'" in name:
                new_name = name.replace("E'", "E^")
                old_path = os.path.join(root, name)
                new_path = os.path.join(root, new_name)
                os.rename(old_path, new_path)
                print(f"Renamed folder: {old_path} to {new_path}")

        for name in files:
            if "e'" in name:
                new_name = name.replace("e'", "e^")
                old_path = os.path.join(root, name)
                new_path = os.path.join(root, new_name)
                os.rename(old_path, new_path)
                print(f"Renamed file: {old_path} to {new_path}")

# Panggil fungsi untuk mengubah nama file atau folder
rename_files_folders(mypath)
```
